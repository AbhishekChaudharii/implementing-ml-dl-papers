```python
import numpy as np
a = np.array([1, 2, 3])      # shape (3,)
b = np.array([4, 5, 6])     # shape (3,)
```

```python
np.outer(a,b)

# np.matmul(a, b) on two 1D vectors computes the dot product (inner product), not the outer product — it gives a scalar 1*4 + 2*5 + 3*6 = 32
```

#### Implement it two ways:

Using nested for loops

Using meshgrid + element-wise multiply (no loops)

```python
# Using nested for loops
a = [1, 2, 3]      # shape (3,)
b = [4, 5, 6]      # shape (3,)
res = []
for i in range(len(a)):
    temp = []
    for j in range(len(b)):
        temp.append(a[i]*b[j])
    res.append(temp)
res
```

```python
# Using numpy which uses broadcasting underneath
a = np.array([1, 2, 3])      # shape (3,)
b = np.array([4, 5, 6])     # shape (3,)

a = a.reshape(3,1) # shape (3,1)
b = b.reshape(1,3) # shape (1,3)

a*b
```

#### No loops implementation - Without using pre-built functions 


```python
a = [1, 2, 3]
b = [4, 5, 6]
# Manually create all (i,j) index combinations
indices = [(i, j) for i in range(len(a)) for j in range(len(b))]

print(indices)
```

```python
# Compute updates
updates = [a[i] * b[j] for i, j in indices]
updates
```

```python
# Scatter into output
res = [[0]*len(b) for _ in range(len(a))]
print(res)
for idx, (i, j) in enumerate(indices):  # (i,j) in this line is tuple unpacking written in short
    print(i,j)
    res[i][j] = updates[idx]
```

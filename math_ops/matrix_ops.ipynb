```python

def my_matmal(mat_a,mat_b):
    """Performs matrix multiplication on Python List"""
    a_rows = len(mat_a)
    a_cols = len(mat_a[0])
    b_rows = len(mat_b)
    b_cols = len(mat_b[0])

    shared_dim = b_rows # or a_cols -> Anything works
    if a_cols!=b_rows:
        raise ValueError(f"Shape mismatch! A has {a_cols} cols but B has {b_rows} rows")
    # output_shape = (a_rows,b_cols)
    res = []
    for i in range(a_rows):
        temp = []
        for j in range(b_cols):
            total = 0
            for k in range(shared_dim):
                total+=mat_a[i][k] * mat_b[k][j]
            temp.append(total)
        res.append(temp)
    
    return res
        
```

```python
# Test 1 — square matrices
A = [[1, 2],
     [3, 4]]
B = [[5, 6],
     [7, 8]]
print(my_matmal(A, B))  # [[19, 22], [43, 50]]

# Test 2 — uneven shapes (3x2) @ (2x3)
A = [[1, 2],
     [3, 4],
     [5, 6]]
B = [[5, 6, 7],
     [7, 8, 9]]
print(my_matmal(A, B))  # [[19, 22, 25], [43, 50, 57], [67, 78, 89]]

# Test 3 — shape mismatch should raise error
A = [[1, 2, 3]]  # 1x3
B = [[1, 2],     # 2x2
     [3, 4]]
print(my_matmal(A, B))  # ValueError!
```

```python


#**  Using the scatter pattern (indices → updates → scatter)
# 1. Create indices with all possible combinations
# 2. Compute values and store them
# 3. Make output matrix of zeros
# 4. Place each update at its correct (i, j) position in the output matrix



# TODO: Vectorized scatter approach


def my_mat_mal_vec(mat_a,mat_b):

    # TODO: 1. Create indices with all possible combinations

    indices = []

    for i in range(len(mat_a)):
        for j in range(len(mat_b[0])):
            temp = []
            for k in range(len(mat_b)):
                temp.append(((i,k),(k,j)))
            indices.append(temp)
    # print(f"Indices: {indices}")

    # TODO: 2. Compute values and store them
    updates = []
    for idx,pairs in enumerate(indices):
        total = 0
        for pair in pairs:
            i,j = pair[0]
            x,y = pair[1]
            total+=mat_a[i][j] * mat_b[x][y]
        updates.append(total)

    # print(f"updates: {updates}")
        
        
    # TODO: 3. Make output matrix of zeros
    a_rows = len(mat_a)
    b_cols = len(mat_b[0])
    output_grid = [[0] * b_cols for _ in range(a_rows)]
    # print(output_grid)

    # TODO: 4. Place each update at its correct (i, j) position in the output matrix
    count = 0
    for outputs in output_grid:
        for output_loc in range(len(outputs)):
            outputs[output_loc] = updates[count]
            count+=1

    # print(output_grid)

    return output_grid

        
```

```python
# Better structured code by Claude
def mat_mal_claude(mat_a,mat_b):
    # CHANGE 1: Flat (i,j,k) triplets instead of nested coordinate pairs
    indices = []
    for i in range(len(mat_a)):
        for j in range(len(mat_b[0])):
            for k in range(len(mat_b)):
                indices.append((i, j, k))  # simple triplet, not ((i,k),(k,j))
    print(f"Indices: {indices}")

    # CHANGE 2: No summing here — just one multiplication per triplet
    updates = []
    for i, j, k in indices:
        updates.append(mat_a[i][k] * mat_b[k][j])  # flat list, no total accumulation
    print(f"Updates: {updates}")

    # SAME: output grid of zeros
    a_rows = len(mat_a)
    b_cols = len(mat_b[0])
    output_grid = [[0] * b_cols for _ in range(a_rows)]

    # CHANGE 3: += instead of = (accumulation happens here, not in step 2)
    for idx, (i, j, k) in enumerate(indices):
        output_grid[i][j] += updates[idx]  # += not =

    return output_grid
```

## ReLU

$$f(x) = \max(0, x)$$

```python


def relu(x):
    return x if x>0 else 0
```

```python
relu(2)
relu([-2, -1, 0, 1, 2]) #fails here 
```

##### To apply ReLU to matrix or array using NumPy maximum function is better

```python
import numpy as np

def relu(z):
    return np.maximum(0,z)
```

```python
print(relu(2))
print(relu(np.array([-2, -1, 0, 1, 2])))
```

## Leaky ReLU

$$f(x) = \begin{cases} x & \text{if } x > 0 \\ \alpha x & \text{if } x \leq 0 \end{cases}$$

```python


def leaky_ReLU(z,leak=0.01):
    return np.maximum(leak*z,z)
```

```python
print(leaky_ReLU(2))
print(leaky_ReLU(np.array([-2, -1, 0, 1, 2])))
```

## ELU - exponential linear unit

$$f(x) = \begin{cases} x & \text{if } x > 0 \\ \alpha(e^x - 1) & \text{if } x \leq 0 \end{cases}$$

```python
def elu(z,alpha=1):
    return np.where(z>0,z,alpha*(np.exp(z)-1))
```

```python
print(elu(2))
print(elu(np.array([-2, -1, 0, 1, 2])))
```

## scaled ELU

$$f(x) = \lambda \begin{cases} x & \text{if } x > 0 \\ \alpha(e^x - 1) & \text{if } x \leq 0 \end{cases}$$
$$\alpha \approx 1.6732, \quad \lambda \approx 1.0507$$

```python
def selu(x, alpha=1.6732, lambda_=1.0507):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
print(selu(2))
print(selu(np.array([-2, -1, 0, 1, 2])))
```

## Gaussian Error Linear Unit (GELU)

$$f(x) = 0.5x\left(1 + \tanh\left(\sqrt{\frac{2}{\pi}}\left(x + 0.044715x^3\right)\right)\right)$$

```python
def GeLU(z):
    return (0.5*z*(1+np.tanh(np.sqrt(2/np.pi)*(z+0.044715*z**3))))
```

```python
print(GeLU(2))
print(GeLU(-2))
print(GeLU(np.array([-2, -1, 0, 1, 2])))
```

## Swish

$$f(x) = \frac{x}{1 + e^{-\beta x}}$$
$$\beta = 1 \text{ (SiLU)}, \quad \beta = 1.702 \text{ (GELU approximation)}$$

```python
def swish(z,beta = 1.702):
    return z*(1 / (1 + np.exp(-beta*z)))
```

```python
print(swish(2))
print(swish(-2))
print(swish(np.array([-2, -1, 0, 1, 2])))
```

## Sigmoid

$$f(x) = \frac{1}{1 + e^{-x}}$$

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

## Tanh

$$f(x) = \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
def tanh(z):
    return np.tanh(z)
```

## Softmax

$$f(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}$$

```python
def softmax(z):
    return np.exp(z)/np.sum(np.exp(z))
```

```python
print(softmax(np.array([-2, -1, 0, 1, 2])))
```

```python
softmax(np.array([1000, 2000, 3000]))  
# can't handle large numbers- overflows
```

#### Numerically stable softmax :

$$f(x_i) = \frac{e^{x_i - \max(x)}}{\sum_{j=1}^{n} e^{x_j - \max(x)}}$$

```python
# Better implementation -> Subtract the max before exponentiating 
# More numerically stable implementation
def softmax(z):
    e_x = np.exp(z - np.max(z)) # Subtract the max from input before exponentiating 
    return e_x/np.sum(e_x)
```

```python
softmax(np.array([1000, 2000, 3000]))  
```

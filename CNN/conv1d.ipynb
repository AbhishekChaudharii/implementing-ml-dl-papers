```python
import numpy as np
```

$$y = x*w \to y[i] = \sum_j x[i+j] . w[K-1-j]$$

w[K-1-j] means flipped kernel

K means length of kernel


```python

def conv1d(x,w):
    """x is the signal and w is the kernel
    Assumes stride=1 and valid padding
    """
    x = np.array(x)
    w = np.array(w)
    flip_kernel = w[::-1].copy()
    n_conv = len(x)-len(w)+1
    res = np.zeros(n_conv)
    for i in range(n_conv):
        ans = x[i:len(w)+i] * flip_kernel
        res[i] = np.sum(ans)
    return res
```

```python
x = [2,3,4]
w = [1,-1]
my_output = conv1d(x,w)
np_conv_output = np.convolve(x,w,mode="valid")
```

```python
np.testing.assert_array_equal(actual=np_conv_output,desired=my_output)
```

```python
x = [1,2,3,0,0]
w = [1,0,-1]
my_output = conv1d(x,w)
np_conv_output = np.convolve(x,w,mode="valid")
```

```python
np.testing.assert_array_equal(actual=np_conv_output,desired=my_output)
```

$$y = x*w \to y[i] = \sum_j x[i+j] . w[j]$$

```python

def cross_conv1d(x,w):
    """x is the signal and w is the kernel
    Assumes stride=1 and valid padding
    """
    x = np.array(x)
    w = np.array(w)
    # flip_kernel = w[::-1].copy() #! remove kernel flip
    n_conv = len(x)-len(w)+1
    res = np.zeros(n_conv)
    for i in range(n_conv):
        ans = x[i:len(w)+i] * w
        res[i] = np.sum(ans)
    return res
```

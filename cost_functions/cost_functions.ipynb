## MSE

$$\text{MSE}(y, \hat{y}) = \frac{\sum_{i=0}^{N - 1} (y_i - \hat{y}_i)^2}{N}$$

```python
def mse(y_true,y_pred):
    return np.mean((y_true - y_pred)**2)
    
```

```python
y_true = np.array([1.0, 2.0, 3.0])
y_pred = np.array([1.1, 1.9, 3.2])

print(mse(y_true, y_pred)) 
```

```python
y_true = np.array([[1.0, 2.0], [3.0, 4.0]])
y_pred = np.array([[1.1, 1.9], [3.2, 3.8]])
print(mse(y_true, y_pred))
```

$$
\text{LogLoss} = -\frac{1}{N} \sum_{i=1}^{N} \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right)
$$


```python
def log_loss(y_true,y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1-y_true)*np.log(1-y_pred))
```

```python
# Binary
y_true = np.array([1, 0, 1, 1])
y_pred = np.array([0.9, 0.1, 0.8, 0.6])
print(log_loss(y_true, y_pred)) 
```

```python
def log_loss(y_true,y_pred):
    """Numerically stable log loss version"""
    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7) 
    return -np.mean(y_true * np.log(y_pred) + (1-y_true)*np.log(1-y_pred))
```

np.log(0)  returns -inf

When applying chain rule in backprop : If 
y_true = 1
y_pred = 0.0   
then -y_true / y_pred = -1 / 0.0 ==> -inf


np.clip limits values of array from [0.0000001, 0.9999999] -> which avoids generating -inf values

-inf becomes nan values

```python
import torch
```

```python
class LinearReg:
    def __init__(self,lr=0.01,iterations=100,random_state = 42):
        self.lr = lr
        self.iterations = iterations
        torch.manual_seed(random_state)
        
    def cost_func(self,y_hat,y_true):
        return torch.mean((y_true - y_hat)**2)
    
    def fit(self,X,y):
        X = torch.tensor(X, dtype=torch.float32)
        y = torch.tensor(y, dtype=torch.float32)

        self.weights = torch.randint(low=1,high=10,size=(X.shape[1],),dtype=torch.float32,requires_grad=True) 
        self.bias = torch.tensor(1,dtype=torch.float32,requires_grad=True)

        for iter in range(self.iterations):
            y_hat = X @ self.weights + self.bias
            loss = self.cost_func(y_hat,y)
            loss.backward() #autograd

            with torch.no_grad():
                # self.weights = self.weights - self.lr * self.weights.grad
                self.weights -= self.lr * self.weights.grad
                # self.bias = self.bias - self.lr * self.bias.grad
                self.bias -= self.lr * self.bias.grad
            
            self.weights.grad.zero_()
            self.bias.grad.zero_()

            if iter%10==0:
                print(f"Iter: {iter} Loss: {loss}")
            





```

```python
import numpy as np

# 100 samples, 3 features
X_train = np.random.randn(100, 3)

# true relationship: y = 2x1 + 3x2 - x3 + noise
y_train = X_train @ np.array([2.0, 3.0, -1.0]) + np.random.randn(100) * 0.1

model = LinearReg()
model.fit(X_train,y_train)
```
